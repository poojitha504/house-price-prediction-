# House Price Prediction - Data Analysis

Objective:
Analyze dataset and understand how area, bedrooms, bathrooms affect price.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("House_data.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.scatter(data['area'], data['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
plt.scatter(data['bedrooms'], data['price'])
plt.show()

plt.scatter(data['bathrooms'], data['price'])
plt.show()
```
